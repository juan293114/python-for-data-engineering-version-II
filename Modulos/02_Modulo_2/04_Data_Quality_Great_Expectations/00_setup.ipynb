```python
from pyspark.sql import functions as F, types as T, Window
import json
import random
import datetime as dt
import os

# Ajusta estos valores a tu entorno
CATALOG = "workspace"
SCHEMA = "gx_lab"
VOLUME = "gx_volume"

# Contexto SQL
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {CATALOG}.{SCHEMA}")
spark.sql(f"USE CATALOG {CATALOG}")
spark.sql(f"USE {SCHEMA}")

# Crear el volumen si no existe
spark.sql(f"CREATE VOLUME IF NOT EXISTS {CATALOG}.{SCHEMA}.{VOLUME}")

# Ruta principal dentro del volumen
BASE_PATH = f"/Volumes/{CATALOG}/{SCHEMA}/{VOLUME}/great_expectations_lab"
VALIDATION_PATH = f"{BASE_PATH}/validations"
QUARANTINE_PATH = f"{BASE_PATH}/quarantine"

# Crear carpetas dentro del volumen
for p in [BASE_PATH, VALIDATION_PATH, QUARANTINE_PATH]:
    os.makedirs(p, exist_ok=True)

def save_json(path: str, payload: dict):
    os.makedirs(os.path.dirname(path), exist_ok=True)
    with open(path, "w", encoding="utf-8") as f:
        json.dump(payload, f, default=str, indent=2, ensure_ascii=False)

def load_json(path: str) -> dict:
    with open(path, "r", encoding="utf-8") as f:
        return json.load(f)

print("Setup OK")
print(f"Catalog: {CATALOG}")
print(f"Schema: {SCHEMA}")
print(f"Volume: {VOLUME}")
print(f"Base path: {BASE_PATH}")
```

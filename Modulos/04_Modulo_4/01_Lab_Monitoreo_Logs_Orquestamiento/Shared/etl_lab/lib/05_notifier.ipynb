```python
import smtplib
from email.message import EmailMessage

def send_failure_email(
    smtp_user: str,
    smtp_app_password: str,
    to_email: str,
    subject: str,
    body: str,
    smtp_host: str = "smtp.gmail.com",
    smtp_port: int = 587
):
    if not smtp_user or not smtp_app_password or not to_email:
        raise ValueError("Faltan parámetros SMTP o email destino")

    msg = EmailMessage()
    msg["From"] = smtp_user
    msg["To"] = to_email
    msg["Subject"] = subject
    msg.set_content(body)

    with smtplib.SMTP(smtp_host, smtp_port, timeout=30) as server:
        server.ehlo()
        server.starttls()
        server.ehlo()
        server.login(smtp_user, smtp_app_password)
        server.send_message(msg)

def try_send_failure_email(**kwargs):
    try:
        send_failure_email(**kwargs)
        return True, None
    except Exception as e:
        return False, str(e)
```

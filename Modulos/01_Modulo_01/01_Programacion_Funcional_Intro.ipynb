# Programación Funcional en Python (Introducción)

**Objetivo:** entender la idea de *transformar datos con funciones* y usar:

- **lambda** → función corta
- **map** → transformar
- **filter** → filtrar
- **reduce** → combinar
- **list** → visualizar resultados

> Nota: Python es multi‑paradigma. Aquí usaremos un estilo funcional *práctico* (no “puro” al 100%).


## 1) ¿Qué es programación funcional?

Es una forma de escribir código donde pensamos en **transformaciones**:

`datos → transformar → filtrar → combinar → resultado`

En vez de mezclar todo en un bloque grande, preferimos **funciones pequeñas** (tipo “bloques de lego”).

### Imperativo (tradicional) vs funcional (declarativo)

- **Imperativo:** control paso a paso con bucles y variables intermedias.
- **Funcional:** describimos qué transformación queremos (map/filter/reduce).


````python
# Ejemplo rápido: duplicar valores
numeros = [1, 2, 3, 4]

# Tradicional (imperativo)
resultado_for = []
for n in numeros:
    resultado_for.append(n * 2)

# Estilo funcional
resultado_map = list(map(lambda x: x * 2, numeros))

resultado_for, resultado_map
````

## 2) lambda → función corta

**lambda** es una forma compacta de definir una función en una sola línea.

**Con `def`:**
```python
def duplicar(x):
    return x * 2
```

**Con `lambda`:**
```python
lambda x: x * 2
```

Se usa mucho cuando la función es pequeña y se va a usar una sola vez.


````python
#Tradicional
def duplicar_def(x):
    return x * 2

duplicar_def(5)
````

````python
#Funcional
duplicar_lambda = lambda x: x * 2

duplicar_lambda(5)
````

## 3) map → transformar

`map(funcion, iterable)` aplica la función a **cada** elemento.

En Python, `map()` devuelve un **iterable** (no una lista). Por eso normalmente usamos `list(...)` para ver el resultado.


````python
numeros = [1, 2, 3, 4]

# Tradicional
cuadrados_for = []
for n in numeros:
    cuadrados_for.append(n * n)

cuadrados_for,
````

````python
numeros = [1, 2, 3, 4]
# Funcional
cuadrados_map = list(map(lambda x: x * x, numeros))

cuadrados_map
````

### Ejemplo más real (data)
Convertir montos en texto a `float`:


````python
montos_txt = ["10.5", "20", "30.2"]
montos = list(map(float, montos_txt))
montos
````

## 4) filter → filtrar

`filter(condicion, iterable)` se queda con los elementos que cumplen la condición.

También devuelve un **iterable**, así que usamos `list(...)` para visualizar.


````python
numeros = [1, 2, 3, 4, 5, 6]

# Tradicional
pares_for = []
for n in numeros:
    if n % 2 == 0:
        pares_for.append(n)

pares_for
````

````python
numeros = [1, 2, 3, 4, 5, 6]

# Funcional
pares_filter = list(filter(lambda x: x % 2 == 0, numeros))

pares_filter
````

## 5) reduce → combinar

`reduce(funcion, iterable)` combina todos los elementos en **un solo resultado**.

Piénsalo así:

- Toma los dos primeros → aplica función
- Usa el resultado con el siguiente
- Repite hasta terminar

Necesitas importarlo:
```python
from functools import reduce
```


````python
from functools import reduce

mayores = [4, 6, 8]

# Tradicional
suma_for = 0
for x in mayores:
    suma_for += x

suma_for
````

````python
from functools import reduce

mayores = [4, 6, 8]

# reduce
suma_reduce = reduce(lambda a, b: a + b, mayores)

suma_reduce
````

### Ver cómo reduce combina paso a paso
Para entenderlo, hacemos una versión “debug”:


````python
def reduce_debug(func, items):
    it = iter(items)
    acumulado = next(it)
    print("Inicio:", acumulado)
    for siguiente in it:
        print(f"Combinar: {acumulado} con {siguiente}")
        acumulado = func(acumulado, siguiente)
        print("Acumulado:", acumulado)
    return acumulado

reduce_debug(lambda a, b: a + b, [4, 6, 8])
````

## 6) list() → visualizar resultados

`map()` y `filter()` no devuelven una lista directamente; devuelven un **iterable**.

`list(...)` convierte ese iterable en una lista para poder verla/imprimirla fácilmente.


````python
numeros = [1, 2, 3, 4]

m = map(lambda x: x * 2, numeros)
print("map devuelve:", m)

print("list(map(...)) devuelve:", list(map(lambda x: x * 2, numeros)))
````

## 7) Mini‑pipeline (map + filter + reduce)

**Caso:**
1) Duplicar montos
2) Quedarnos con los > 3
3) Sumarlos


````python
from functools import reduce

numeros = [1, 2, 3, 4]

dobles = list(map(lambda x: x * 2, numeros))
mayores = list(filter(lambda x: x > 3, dobles))
suma = reduce(lambda a, b: a + b, mayores)

print("Dobles:", dobles)
print("Mayores:", mayores)
print("Suma:", suma)
````

# Ejercicios (con solución)

## Ejercicio 1
Dada la lista `nums = [1,2,3,4,5]`, obtener los **cuadrados de los pares**.


````python
nums = [1, 2, 3, 4, 5]

# Solución tradicional
res_for = []
for n in nums:
    if n % 2 == 0:
        res_for.append(n * n)

# Solución funcional
pares = filter(lambda x: x % 2 == 0, nums)
res_func = list(map(lambda x: x * x, pares))

res_for, res_func
````

## Ejercicio 2
Dada una lista de montos `montos = [50, 120, 30, 200]`, aplicar **IGV (18%)** y quedarte solo con los mayores a **150**.


````python
montos = [50, 120, 30, 200]

# Solución tradicional
res_for = []
for m in montos:
    m_igv = m * 1.18
    if m_igv > 150:
        res_for.append(m_igv)

# Solución funcional
con_igv = map(lambda x: x * 1.18, montos)
res_func = list(filter(lambda x: x > 150, con_igv))

res_for, res_func
````

## Ejercicio 3
Dada la lista `vals = ["10", "20", "error", "30"]`, convertir a enteros y sumar los válidos.

**Nota:** aquí usaremos una función auxiliar para manejar el error.


````python
from functools import reduce

vals = ["10", "20", "error", "30"]

def to_int_or_none(s):
    try:
        return int(s)
    except (TypeError, ValueError):
        return None

# Solución funcional por etapas
convertidos = map(to_int_or_none, vals)
validos = filter(lambda x: x is not None, convertidos)
suma = reduce(lambda a, b: a + b, validos, 0)  # 0 como acumulador inicial

print("Suma válidos:", suma)
````

```python
import pandas as pd

possible_keys = ["messages", "transcript", "conversation", "turns"]
message_rows = None

for key in possible_keys:
    value = data.get(key)
    if isinstance(value, list):
        message_rows = value
        print(f"Encontrado detalle en la llave: {key}")
        break

if message_rows:
    df_messages = pd.json_normalize(message_rows)
    display(df_messages)
else:
    print("No se encontró una lista de mensajes en el JSON.")
    print("Llaves disponibles:", list(data.keys()))
```
